## ¿ Y que es un objeto?

Un objeto es “una cosa”. Y, si una cosa es un sustantivo, entonces un objeto es un sustantivo.

## Las cualidades de un objeto

Describir un objeto, es simplemente mencionar sus cualidades. Las cualidades son
adjetivos.

Para describir “la manera de ser” de un objeto, debemos preguntarnos **¿cómo es el objeto?**

Toda respuesta que comience por **“el objeto es”**, seguida de un adjetivo será una cualidad del objeto.

Además de cualidades, los objetos “tienen
otras cosas”. Estas “otras cosas”, son aquellas “pseudo-cualidades” que en vez de
responder a **¿cómo es el objeto?** responden a **“¿cómo está compuesto el objeto?”** o
incluso, aún más simple **“¿Qué tiene el objeto?”**.

## Accciones de un objeto

Los objetos tiene la capacidad de realizar acciones.Las acciones, son verbos.


Es decir, que para conocer las capacidades de un objeto, debes preguntarte **“¿Qué puede hacer
el objeto?”**  y la respuesta a esta pregunta, estará dada por todas aquellas que comiencen por la frase **“el objeto puede”** seguida de un verbo en infinitivo.

En lugar de escribir muchas declaraciones if..else, puedes usar la declaración de match.



```
match expression:
  case x:
    code block
  case y:
    code block
  case z:
    code block
```



````python
dia = 4

if dia==1:
  print("Lunes")
elif dia==2:
  print("Martes")
elif dia==3:
  print("Miercoles")
elif dia==4:
  print("Jueves")
elif dia==5:
  print("Viernes")
elif dia==6:
  print("Sabado")
elif dia==7:
  print("Domingo")
else:
  print("Día no valido")
````

````python
dia = 4
match dia:
  case 1:
    print("Lunes")
  case 2:
    print("Martes")
  case 3:
    print("Miercoles")
  case 4:
    print("Jueves")
  case 5:
    print("Viernes")
  case 6:
    print("Sabado")
  case 7:
    print("Domingo")
  case _:
    print("Día no valido")
````

````python
dia = 3
match dia:
  case 1 | 2 | 3 | 4 | 5:
    print("Día entre semana")
  case 6 | 7:
    print("Fin de semana!")
````

## While

Con el bucle while podemos ejecutar un conjunto de instrucciones siempre que una condición sea verdadera.

````python
i = 1
while i < 10:
  print(i)
  break
````

Con la declaración break podemos detener el bucle incluso si la condición while es verdadera:

````python
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
````

Con la instrucción continue podemos detener la iteración actual y continuar con la siguiente:

````python
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
````

````python
i = 1
while i < 6:
  print(i)
  i += 1
else:
  print("i ya es menor que 6")
````
