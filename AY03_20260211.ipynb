## F-String

Para especificar una cadena como una f-string, solo hay que colocar una f delante del literal de cadena y agregue llaves {} como marcadores de posición para variables y otras operaciones.

````python
edad = 24

# sin f-string
print("Yo tengo", edad, "años")

# con f-string
print(f"Yo tengo {edad} años")
````

tambien se pueden hacer operaciones dentro los marcadores de posición:

````python
print(f"El precio de ese producto es de: {5*18} pesos mexicanos")
````

## Estructuras de control


### IF - ELSE

La instrucción if evalúa una condición (una expresión cuyo resultado es Verdadero o Falso ). Si la condición es verdadera, se ejecuta el bloque de código dentro de la instrucción if. Si la condición es falsa, se omite el bloque de código.

Es como decirle a Python *al llegar a este punto, ejecuta esta(s) acción(es) sólo so esta condición es cierta*

```
if condición:
      acción
```

````python
a= 1
b= 2

if b > a:
  print("b es mayor que a")
````

### Identación

La sangría se refiere a los espacios al principio de una línea de código.

Mientras que en otros lenguajes de programación la sangría en el código es solo para la legibilidad, la sangría en Python es muy importante.

Python utiliza sangría para indicar un bloque de código.

Las variables booleanas se pueden utilizar directamente en declaraciones if sin operadores de comparación

````python
variable = True
if variable:
  print("Hello World!")
````

### Else

La palabra else significa sino o en caso contrario

````python
a= 2
b= 1

if b < a:
  print("b es menor que a")
else:
  print("b es mayor que a")
````

````python
x=4

if(x%2 == 0):
    print("entera")
else:
    print("El reciduo es distinto de cero")
````

### elif

La palabra clave elif le permite verificar múltiples expresiones para verificar si son Verdaderas y ejecutar un bloque de código tan pronto como una de las condiciones se evalúe como Verdadero  y omite las demás.

Solo se ejecutará la primera condición verdadera. Incluso si varias condiciones son verdaderas, Python se detiene después de ejecutar el primer bloque coincidente.

Es recomendable usar elif cuando tenga que verificar varias condiciones mutuamente excluyentes.Ya que esto es más eficiente que usar varias sentencias if independientes , ya que Python detiene la verificación al encontrar una condición verdadera.

````python
edad = 24

if edad < 10:
  print("Eres un niño")
elif edad < 20:
  print("Eres un adolescente")
elif edad < 60:
  print("Eres un adulto")
elif edad >= 60:
  print("Eres un adulto mayor")
````

### **Slicing**
Se realiza definiendo los valores de índice del primer elemento y el último elemento de la lista principal que se requiere en la lista dividida. Se escribe como [ a : b ] donde a,b son los valores de índice de la lista padre. Si a o b no está definido, entonces el valor del índice se considera el primer valor para a si a no está definido y el último valor para b cuando b no está definido.
