## Ejercicio

Escribe un prigrama que reciba el radio de un círculo y le despliegue al usuario el siguiente menú:

    A. Calcular el diámetro
    B. Calcular el perímetro
    C. Calcular el área

De acuerdo a la opción del resultado, que el programa seleccione el procedimiento y devuelva el resultado.

```python
from math import pi

radio = float(input("Ingresa el radio de tu circulo: "))

opcion = input("¿Qué quieres calcular? \n A = Diametro \n B = Perimetro \n C = Área \n")

opcion = opcion.capitalize()

if opcion == 'A':
  diametro = 2*radio
  print("El diametro es: ", diametro)
elif opcion == 'B':
  perimetro = pi * (2*radio)
  print("El perimetro es: ", perimetro)
elif opcion == 'C':
  area = pi * (radio**2)
  print("El area es: ", area)
else:
  print("Opción no valida")
```

## Ejericicio: Diseccionando un número

Desarrolla un programa que lea un número entero de cuatro dígitos y muestre la suma de sus dígitos (ej. si ingresa 3141, debe mostrar 3+1+4+1 = 9).

```python
# Método 1:

numero = int(input("Ingresa un número entero de 4 dijitos "))

d1 = numero // 1000
numero %= 1000

d2 = numero // 100
numero %= 100

d3 = numero // 10
numero %= 10

d4 = numero // 1

suma = d1 + d2 + d3 + d4

print("La suma de los digitos es: ", suma)
```

```python
# Método 2:

numero = int(input("Ingresa un número entero de 4 dijitos "))

d1 = numero // 1000
numero = numero - 1000 * d1

d2 = numero // 100
numero = numero - 100 * d2

d3 = numero // 10
numero = numero - 10 * d3

d4 = numero // 1

suma = d1 + d2 + d3 + d4

print("La suma de los digitos es: ", suma)


```

## Ejercicio: Unidades de tiempo

Pide al usuario que ingrese una duración en Días, Horas, Minutos y Segundos (4 inputs separados). Calcula el total de todo eso en segundos puros.

Pide al usuario una cantidad masiva de segundos (ej. 100,000) y conviértela al formato D:HH:MM:SS.
Lo que evalúa:

```python
segundos_totales = int(input("Ingresa una cantidad grandota de segundos: "))

segundos_minuto = 60
segundos_hora = 3600
segundos_dia = 86400

dias = segundos_totales // segundos_dia
segundos_totales %= segundos_dia

horas = segundos_totales // segundos_hora
segundos_totales %= segundos_hora

minutos = segundos_totales // segundos_minuto


segundos = segundos_totales % segundos_minuto

print(f"{dias}:{horas}:{minutos}:{segundos}")
```

## Tuplas

Una tupla es una colección ordenada e inmutable (lo que significa que no podemos cambiar, agregar o eliminar elementos después de que se haya creado la tupla.)

Las tuplas se escriben con corchetes.


```python
materias=('calculo', 'algebra', 'computacion', 'fisica','geometria')
```

```python
print(materias)
```

```python
print(type(materias))
```

```python
print(materias[0])
```

```python
print(materias[-1])
```

```python
print(materias[1:3])
```

Las tuplas son similares a las listas, pero la única gran diferencia es que los elementos dentro de una lista se pueden cambiar, pero en la tupla no se pueden cambiar.

```python
materias[0] = "literatura"
```

```python
tupla1 = () # Se define una tupla vacia
tupla2 = (tupla1, "sol", 2.17, 1) #Esta tupla no está vacía y sus elementos están separados por comas.
tupla3 = tuple() #Esta es la otra forma de definir una tupla a partir de la función tuple()
```

```python
tupla1 = ("manzana")
print(type(tupla1))
```

Cambiar Valores De Tuplas

Una vez que se crea una tupla, no se pueden cambiar sus valores. Las tuplas son inmutables, o inmutables, como también se les llama.

Pero hay una solución. Puedes convertir la tupla en una lista, cambiar la lista y convertir la lista de nuevo en una tupla.

```python
frutas_tupla = ("manzana", "platano", "cereza")
print(type(frutas_tupla))
```

```python
frutas_lista = list(frutas_tupla)
print(type(frutas_lista))
```

```python
frutas_lista[1] = "kiwi" # Cambiamos el elemento con indice 1(platano) por kiwi

frutas_tupla = tuple(frutas_lista)

print(frutas_tupla)
```
