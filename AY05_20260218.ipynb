##Ejercicio:  Maquina de cambio

Escribe un programa que lea una cantidad entera de dinero (ej. 875 pesos). El programa debe calcular cuántos billetes de 500, 200, 100, 50, 20, y monedas de 10, 5, 2 y 1 peso. se necesitan para dar esa cantidad usando el mínimo número de billetes/monedas posible.

```python
print("--- Maquina de cambio ---")

cantidad = int(input("Ingresa la cantidad de cambio a dar (en pesos enteros): $"))

print(f"\nPara entregar ${cantidad}, necesitas:")


billetes_500 = cantidad // 500
cantidad %= 500  # Esto es equivalente a: cantidad = cantidad % 500

billetes_200 = cantidad // 200
cantidad %= 200

billetes_100 = cantidad // 100
cantidad %= 100

billetes_50 = cantidad // 50
cantidad %= 50

billetes_20 = cantidad // 20
cantidad %= 20

monedas_10 = cantidad // 10
cantidad %= 10

monedas_5 = cantidad // 5
cantidad %= 5

monedas_2 = cantidad // 2
cantidad %= 2

monedas_1 = cantidad // 1
cantidad %= 1

if billetes_500 > 0:
    print(billetes_500, "billete(s) de $500")

if billetes_200 > 0:
    print(f"{billetes_200} billete(s) de $200")

if billetes_100 > 0:
    print(f"{billetes_100} billete(s) de $100")

if billetes_50 > 0:
    print(f"{billetes_50} billete(s) de $50")

if billetes_20 > 0:
    print(f"{billetes_20} billete(s) de $20")

if monedas_10 > 0:
    print(f"{monedas_10} moneda(s) de $10")

if monedas_5 > 0:
    print(f"{monedas_5} moneda(s) de $5")

if monedas_2 > 0:
    print(f"{monedas_2} moneda(s) de $2")

if monedas_1 > 0:
    print(f"{monedas_1} moneda(s) de $1")
```

## Ejercicio: Calculo de recibo de un restaurante

Pide el costo de una comida. Calcula el IVA (16%) y la propina (10% sobre el costo sin IVA). Imprime el desglose: Costo original, IVA, Propina y Total.

```python
costo_comida = float(input("Ingresa el costo de la comida: $ "))

iva = costo_comida*0.16
propina = costo_comida*0.1
total = costo_comida + iva + propina

print("----- Desglose de la cuenta ------ ")
print(f"Costo original: ${costo_comida:.2f}")
print(f"IVA(16%):       ${iva:.2f}")
print(f"Propina(10%):   ${propina:.2f}")
print(f"TOTAL A PAGAR:  ${total:.2f}")
```

## Ejercicio: Ordenando sin listas


Lee tres enteros del usuario. Imprímelos ordenados de menor a mayor usando solo las funciones min(), max() y aritmética básica (suma/resta). ¡No se vale usar listas ni if/else!

```python
a = int(input("Ingresa el primer número: "))
b = int(input("Ingresa el segundo número: "))
c = int(input("Ingresa el tercer número: "))

menor = min(a, b, c)
mayor = max(a, b, c)

medio = (a + b + c) - menor - mayor

print(f"Los números ordenados de menor a mayor son: {menor}, {medio}, {mayor}")
```
